```python
import numpy as np

sizes = [100, 200, 400, 800, 1200, 1600]

for n in sizes:
    A = np.random.rand(n, n)
    B = np.random.rand(n, n)

    with open(f'matrix_A_{n}.txt', 'w') as f:
        f.write(f"{n}\n")
        for i in range(n):
            for j in range(n):
                f.write(f"{A[i][j]:.6f} ")
            f.write("\n")

    with open(f'matrix_B_{n}.txt', 'w') as f:
        f.write(f"{n}\n")
        for i in range(n):
            for j in range(n):
                f.write(f"{B[i][j]:.6f} ")
            f.write("\n")

    print(f"Созданы матрицы {n}x{n}")
```

```python
import matplotlib.pyplot as plt

sizes = [100, 200, 400, 800, 1200, 1600]
times = [0.008, 0.066, 0.751, 7.267, 49.583, 119.912]

plt.plot(sizes, times, 'ro-')
plt.xlabel('Размер матрицы')
plt.ylabel('Время (с)')
plt.title('Умножение матриц')
plt.grid(True)
plt.savefig('graph.png')
plt.show()
```
